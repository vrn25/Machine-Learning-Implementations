```
import numpy as np
```

```
def hypothesis(theta, X):
  # hypothesis for linear regression
  return np.dot(theta, X)
```

```
def cost_function(theta, X, y):
  # mean sqauared error cost function
  m = X.shape[1]
  return np.sum((hypothesis(theta, X) - y)**2)/(2*m)
```

```
def cost_function_derivative(theta, X, y):
  # cost function derivative with respect to theta
  m = X.shape[1]
  return np.dot((hypothesis(theta, X) - y), X.T)/m
```

```
def gradient_descent(theta, X, y, num_itrs, alpha):
  # updating theta by gradient descent
  m = X.shape[1]
  for i in range(num_itrs):
    theta = theta - alpha * cost_function_derivative(theta, X, y)
    print('Cost function after %d iterations is: %f' %(i+1, cost_function(theta, X, y)))
  return theta
```

```
def linear_regression(theta, X, y, num_iters, alpha):
  return gradient_descent(theta, X, y, num_iters, alpha)
```

```
# randomly initializing feature and parameter values
np.random.seed(7)
num_examples = m = 100
num_features = n = 5
theta = np.random.rand(1, n)
X = np.random.rand(n, m)
y = np.random.rand(1, m) * 5
alpha = 0.05
num_iters = 1000
print('Initial theta values: %s' %theta)
theta = linear_regression(theta, X, y, num_iters, alpha)
print('Final theta values: %s' %theta)
```
